```python
import tweepy as tw
```

```python
accesstoken=config.get('twitter','accesstoken')
```

```python
accesstokensecret=config.get('twitter','accesstokensecret')
apikey=config.get('twitter','apikey')
apisecretkey=config.get('twitter','apisecretkey')
```

```python
auth=tw.OAuthHandler(apikey,apisecretkey)
```

```python
auth.set_access_token(accesstoken,accesstokensecret)
```

```python
api=tw.API(auth,wait_on_rate_limit=True)
```

```python
search_word='#COVID-19 vaccine' 
date_since='2021-05-28'
```

```python
tweets=tw.Cursor(api.search,q=search_word,lang='en',since=date_since).items(100)
```

```python
users_locs = [[tweet.created_at,
               tweet.author.screen_name,
               tweet.author.name,
               tweet.text,
               tweet.retweet_count,
               tweet.favorite_count,
               tweet.user.location,
               tweet.user.followers_count,
               tweet.user.friends_count] for tweet in tweets]
```

```python
users_locs
```

```python
import pandas as pd
tweet_df=pd.DataFrame(data=users_locs,columns=['date','user','name','text','retweet','favorite','location','followers','friends'])
pd.set_option('max_colwidth',800)
tweet_df.head(20)
```

```python
tweet_df.to_csv('/content/test_data_withfollowers.csv')
```
